## Using Python

You can run Python several different ways: 
1. Python interpreter 
2. Python script
3. iPython 



### Python interpreter 

To launch the Python interpreter, type the following from the command line:

    $ python
    

<img src="./images/python4.png">


Try: 

    $ print "hello world"
   
    $ 2 + 3
    

To exit: 

    $ exit()



### Simple Python script

Using a text editing tool of your choice, create a new file with the extension `".py"`. 
- On the VDI, there are several text editors available under the `Applications` --> `Accessories` menu.
- Or alternatively, you can create the file directly from command line using `nano` or `vi`, etc.

    $ nano helloworld.py 




Add some simple lines of Python code to the file and close:

    print("Hello world!")
    print("This is my first Python script.")


From command line, run the script:

    $ python helloworld.py


<img src="./images/python3.png">




### iPython

To launch iPython on the VDI (without having to build locally), the iPython module must first be loaded:

    $ module load ipython/4.2.0-py2.7
    
    
Next, type:

    $ ipython
